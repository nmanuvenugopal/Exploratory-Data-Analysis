```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data from CSV to Dataframe

df = pd.read_csv('winequalityN.csv')
df.head()
```

```python
# Simple descriptive analysis
# Only the numeric column will be taken into consideration

df.describe()
```

```python
# Information about Dataframe columns

df.info()
```

```python
# increased the number of bins to make graph more granular and descriptive

df.hist(column='alcohol', bins = 15)
```

```python
# Doing more cutomization on above graph

df.hist(column='alcohol', bins=15, grid=False, color='Red', figsize=(10,6))
```

```python
# For better stylish graphs we use Seaborn library

import seaborn as sns
```

```python
# In this grapgh kde is true (default value), thats why we have line over the bars.

sns.distplot(df['alcohol'], bins=15)
```

```python
# In this case kde is made to false.

sns.distplot(df['alcohol'], kde=False, bins=15)
```

```python
# Add additional styling and labelling
# Seaborn can leverages some of the Matplotlib libraries

sns.distplot(df['alcohol'], bins=15)
plt.xlabel('Alcohol Percentage')
plt.ylabel('Count')
plt.title('Alcohol percentage in wines')
plt.show()
```

```python
# If we want to change the background of the graph.
# for that we use set_style() method.

sns.set_style(style='dark')
sns.distplot(df['alcohol'], bins=15)
```

```python
# Scatter plots
# The trend shows that as the alcohol content increases acidity decreases

sns.lmplot(x='alcohol' , y='fixed acidity', data=df)
```

```python
# If we doesn't want the regression line in graph

sns.lmplot(x='alcohol', y='fixed acidity', data=df, fit_reg=False)
```

```python
# we can plot 3rd argument as hue argument
# this indicates that redwine always have more chlorides than white wine

sns.lmplot(x='alcohol', y ='chlorides', hue='type', data=df, fit_reg=False)
```

```python
# Boxplots and Outliers

sns.boxplot(data=df['alcohol'])
```

```python
df_offices =  pd.read_csv('dm_office_sales.csv')
df_offices.head()
```

```python
# Scatterplot is used to show the relationship between two continous features or columns.

sns.scatterplot(x='salary', y='sales', data=df_offices)
```

```python
# if we want to alter the graph size and increase its clarity:

plt.figure(figsize=(6,3), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df_offices)
```

```python
# hue paramter will color the plot values based on the values of the column.
# in this case we have chosen a categorical column for hue.

sns.scatterplot(x='salary', y='sales', data=df_offices, hue='level of education')
```

```python
# if the continous column is fed to hue

sns.scatterplot(x='salary', y='sales', data=df_offices, hue='salary')
```

```python
sns.scatterplot(x='salary', y='sales', data=df_offices, hue='level of education', palette='Dark2')
```

```python
# Size variable to show bigger size for bigger datapoint values
# Smaller size for smaller datapoint values.

plt.figure(figsize=(10,5))
sns.scatterplot(x='salary', y='sales', data=df_offices,
                size='salary', hue='level of education')
```

```python
# "S" argument is used to increse the size of all datapoints.

plt.figure(figsize=(10,6))
sns.scatterplot(x='salary', y='sales', data=df_offices, hue='level of education', s=200)
```

```python
# Style argument is used when the different datapoints need to be different.

plt.figure(figsize=(10,6))
sns.scatterplot(x='salary', y='sales', data=df_offices,
                hue='level of education', style='level of education', s=150)
```

```python
# Rugplot
# If plotted against salary it will create single tick for every salary.

sns.rugplot(x='salary', data=df_offices, height=0.5)
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
# Simple boxplot for a continous variable

sns.boxplot(data=df, y='math score')
```

```python
# boxplot showing the distribution of maths score wrt Parental level of education

sns.boxplot(x='parental level of education', y='math score', data=df, orient='v')
```

```python
# Plotting the sub categories using boxplot

sns.boxplot(data=df, x='parental level of education', y='math score', 
            hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
# Violinplots

sns.violinplot(data=df, x='parental level of education', y='math score',
               hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05, 0.5))
```

```python
#jointplot is much similar to scatter plot

sns.jointplot(data=df, x='math score', y='reading score', hue='gender')
```

```python
sns.pairplot(data=df)
```

```python
sns.catplot(data=df, x='lunch', y='math score',hue='test preparation course' ,kind='box')
```

```python
sns.catplot(data=df, x='gender',y='math score', kind='box',
            row='test preparation course')
```

```python
sns.catplot(data=df, x='gender', y='math score', row='lunch',col='test preparation course',
            kind='box')
```

```python
# Lets visualize how PairGrid plots work
# pairplot is internally calling PairGrid to form its plot.

g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g = g.map_diag(sns.histplot)
g = g.map_lower(sns.kdeplot)
```

```python
df_con = pd.read_csv('country_table.csv')
df_con = df_con.set_index('Countries')
df_con
```

```python
# Matrix plot
# It can be divided into HeatMap and ClusterMap

sns.heatmap(data=df_con)
```

```python
# It is clear from the heatmap that Life expectancy column has no significance

sns.heatmap(df_con.drop('Life expectancy', axis=1), linewidths=0.5, annot=True,
            cmap='viridis')
```

```python
# ClusterMap

sns.clustermap(df_con.drop('Life expectancy', axis=1), linewidths=0.5, annot=True,
               cmap='viridis')
```
